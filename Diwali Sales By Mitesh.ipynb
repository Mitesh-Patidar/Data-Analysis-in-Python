# Exploratory Data Analysis

## 1) Gender

```python
ax = sns.countplot(x='Gender',data= df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen= df.groupby(['Gender'], as_index = False) ['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender',y='Amount', data=sales_gen)
```

#### Based on the data presented in both graphs, it is evident that females have placed more orders and spent more money than males.

## 2) Age

```python
ax= sns.countplot (data= df, x='Age Group', hue= 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending= False)

sns.barplot(x='Age Group', y='Amount', data= sales_age)
```

#### From the above graphs, we can observe that the majority of buyers belong to the 26-35 age group, with females being the predominant purchasers.

## 3) State

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(x= 'State', y='Orders',data= sales_state)
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(x='State',y='Amount',data=sales_state)
```

#### From the above graphs, it is evident that the highest number of orders and the highest amount spent originate from Uttar Pradesh, Maharashtra, and Karnataka. However, Kerala ranks 8th in terms of the highest number of orders, whereas Haryana ranks 8th in terms of the highest amount spent. Additionally, Kerala does not rank among the top 10 in terms of the highest amount spent; instead, Bihar is included in the top list.

## 4) Marital Status

```python
ax= sns.countplot(x='Marital_Status',data=df)
sns.set(rc={'figure.figsize':(2,6)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending= False)
sns.set(rc={'figure.figsize':(5,4)})
sns.barplot(x='Marital_Status',y='Amount',hue='Gender', data= sales_state)
```

#### From the above graphs, it is apparent that the majority of buyers are married women, and they have also spent the most.

## 5) Occupation

```python
ax=sns.countplot(x='Occupation', data= df)
sns.set(rc={'figure.figsize':(25,6)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index= False)['Amount'].sum().sort_values(by='Amount',ascending= False)
sns.set(rc={'figure.figsize':(20,6)})
sns.barplot(x='Occupation',y='Amount',data = sales_state)
```

#### From the above graphs, we can observe that individuals employed in the IT, healthcare, and aviation sectors have placed most orders and spent the highest amounts of money, respectively.

## 6) Products Category

```python
ax=sns.countplot(data= df, x='Product_Category')
sns.set(rc={'figure.figsize':(25,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category', y='Amount', data= sales_state)
```

#### From the above graphs, we can see that the highest number of sold products are from the Clothing & Apparel, Food, and Electronics categories. However, the categories that generated the most revenue are Food, Clothing & Apparel, and Electronics.


```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,7)})
sns.barplot(x='Product_ID',y='Orders',data= sales_state)
```

# Conclusion
